```python
from tkinter import *
import time
import datetime as dt
from tkinter import Tk, Label, StringVar, Entry, Button
import winsound as ws

# Function to trigger the alarm
def alarm(setAlarmTimer):
    while True:
        time.sleep(1)
        actualTime = dt.datetime.now()
        currentTime = actualTime.strftime("%H : %M : %S")
        currentDate = actualTime.strftime("%d / %m / %Y")
        the_message = "Current Time: " + str(currentTime)
        print(the_message)
        if currentTime == setAlarmTimer:
            ws.PlaySound("sound.wav", ws.SND_ASYNC)
            break

# Function to get the alarm time from user input
def getAlarmTime():
    alarmSetTime = f"{hour.get()} : {minute.get()} : {second.get()}"
    alarm(alarmSetTime)

# Creating the GUI window
guiWindow = Tk()
guiWindow.title("The Alarm Clock")
guiWindow.geometry("464x200")
guiWindow.config(bg="#87BDD8")
guiWindow.resizable(width=False, height=False)

# Displaying a reminder message about the time format
timeFormat = Label(
    guiWindow,
    text="24-hour format!",
    fg="white",
    bg="#36486B",
    font=("Arial", 15)
).place(
    x=0,
    y=160
)

# Labels for the input fields
add_time = Label(
    guiWindow,
    text="Hour     Minute     Second",
    font=60,
    fg="white",
    bg="#87BDD8"
).place(
    x=220,
    y=10
)

# Label for setting the alarm
setAlarm = Label(
    guiWindow,
    text="Set Time for Alarm: ",
    fg="white",
    bg="#034F84",
    relief="solid",
    font=("Helevetica", 13, "bold")
).place(
    x=5,
    y=50
)

# Input fields for hour, minute, and second
hour = StringVar()
minute = StringVar()
second = StringVar()

hourTime = Entry(
    guiWindow,
    textvariable=hour,
    bg="#FEFBD8",
    width=4,
    font=(20)
).place(
    x=220,
    y=53
)
minuteTime = Entry(
    guiWindow,
    textvariable=minute,
    bg="#FEFBD8",
    width=4,
    font=(20)
).place(
    x=297,
    y=53
)
secondTime = Entry(
    guiWindow,
    textvariable=second,
    bg="#FEFBD8",
    width=4,
    font=(20)
).place(
    x=390,
    y=53
)

# Button to set the alarm
submit = Button(
    guiWindow,
    text="Set The Alarm",
    fg="white",
    bg="#82B74B",
    width=15,
    command=getAlarmTime,
    font=(20)
).place(
    x=140,
    y=100
)

# Running the GUI loop
guiWindow.mainloop()
```
